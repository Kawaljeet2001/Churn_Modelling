```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing the dataset

dataset = pd.read_csv("Churn_Modelling.csv")
dataset
```

# OUR AIM IS TO PREDICT WHETHER THE CUSTOMERS WILL LEAVE THE BANK IN THE FUTURE OR NOT/ THEY EXITED OR NOT

```python
X = dataset.iloc[: , 3:13]
y = dataset.iloc[: , 13]
```

```python
y
```

```python
# Now create fummy variables

geography = pd.get_dummies(X['Geography'] , drop_first = True)
gender = pd.get_dummies(X['Gender'] , drop_first = True)
```

```python
geography
```

```python
#now I will concatinate the geography and gender dataframes with my original X dataframe
```

```python
X = pd.concat([X , geography , gender] , axis = 1)
```

```python
X
```

```python
# now dropping the older columns

X = X.drop(['Geography' , 'Gender'] , axis = 1)
```

```python
X
```

```python
#Now splitting the dataset in Training Set and Test Set

from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = 0)
```

```python
X_test
```

```python
#We will do some feature scaling in Deep Learning as well
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Lets Start Making the ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU , PReLU , ELU
from keras.layers import Dropout
```

```python
#Initializing the ANN

classifier = Sequential()
```

```python
#This will make an empty hidden layer
```

```python
#Adding the second hidden layer
classifier.add(Dense(units = 6 , kernel_initializer = "he_uniform" , activation = "relu" , input_dim = 11))
classifier.add(Dense(units = 6 , kernel_initializer = "he_uniform" , activation = "relu"))
```

```python
#Adding the Output Layer

classifier.add(Dense(units = 1 , kernel_initializer = "glorot_uniform" , activation = "sigmoid"))
```

```python
#Compiling the ANN

classifier.compile(optimizer = "adam" , loss = "binary_crossentropy" , metrics = ['accuracy'])
classifier.summary()
```

```python
#Fitting the Deep Learning Model

model_history = classifier.fit(X_train , y_train , validation_split = 0.33 , batch_size = 10 , epochs = 100 , verbose = 0)
```

```python
model_history
```

```python
print(model_history.history.keys())
```

```python
print(model_history.history['val_accuracy'])
```

```python
#PRedicting the Test SEt Results

y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)

#Making the Confusioin Matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test , y_pred)
```

```python
print(cm)
```

```python
#Calculate the Accuracy Score

from sklearn.metrics import accuracy_score

score = accuracy_score(y_pred , y_test)
```
